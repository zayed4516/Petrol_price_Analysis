```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/drive/MyDrive/today/petrol_price.csv')
```

```python
df.head()
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
avarage_price=df.drop(['date'],axis=1).mean(axis=0).reset_index()
avarage_price.columns=['State','Avarage']
```

```python
avarage_price.head()
```

```python
highest_avg_sort=avarage_price.sort_values(by='Avarage',ascending=False)
highest_avg_sort.head()
```

```python
highest_avg_state=highest_avg_sort.iloc[0]
```

```python
highest_avg_state
```

```python
fig=px.bar(highest_avg_sort,x='State',y='Avarage',title='Highest Average State over the last 5 year')
fig.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig.show()
```

```python
df_maharashtra_delhi=df[['date','Maharashtra','Delhi']]
df_maharashtra_delhi.head()
```

```python
df_maharashtra_delhi=pd.melt(df_maharashtra_delhi,id_vars='date',var_name='State',value_name='Price')
df_maharashtra_delhi.head()
```

```python
fig=px.line(df_maharashtra_delhi,x='date',y='Price',color='State',title='Maharashtra vs Delhi Petrol Price')
fig.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig.show()
```

```python
df_maharashtra_Gujarat=df[['date','Maharashtra','Gujarat']]
df.head()
```

```python
df_maharashtra_Gujarat=pd.melt(df_maharashtra_Gujarat,id_vars='date',var_name='State',value_name='Price')
df_maharashtra_Gujarat.head()
```

```python
fig2=px.line(df_maharashtra_Gujarat,x='date',y='Price',color='State',title='Maharashtra vs Gujarat Petrol Price')
fig2.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig2.show()
```

```python
fig3=px.line(df,x='date',y=df.columns[1:],title='Petrol Price Trend of all States over last 5 years')
fig3.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig3.show()
```

```python
fig4=px.box(df,x='date',y=df.columns[1:],title='Petrol Price Comparision of all States over last 5 years')
fig4.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig4.show()
```

```python
fig5=px.scatter(df,x='date',y=df.columns[1:],title='State Specific Fluctuations in petrol price')
fig5.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig5.show()
```

```python
numeric_columns=df.drop(['date'],axis=1)
numeric_columns.head()
```

```python
correlation_matrix=numeric_columns.corr()
correlation_matrix
```

```python
fig6=px.imshow(correlation_matrix,title='Correlation Matrix')
fig6.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig6.show()
```

```python
variance_prices=df.drop(['date'],axis=1).var(axis=0).reset_index()
variance_prices.columns=['State','Variance']
variance_prices.head()
```

```python
fig7=px.bar(variance_prices,x='State',y='Variance',title='Variance of Petrol Price')
fig7.update_layout(plot_bgcolor='black',paper_bgcolor='black',font=dict(color='white'),xaxis=dict(linecolor='white'),yaxis=dict(linecolor='white'))
fig7.show()
```
